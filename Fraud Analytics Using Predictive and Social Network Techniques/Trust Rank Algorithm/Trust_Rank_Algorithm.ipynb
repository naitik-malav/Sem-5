```python
def getNodes(df):
    nodes = []
    for i in range(len(df)):
        u = df.iloc[i, 0]
        v = df.iloc[i, 1]

        if u not in nodes:
            nodes.append(u)
        if v not in nodes:
            nodes.append(v)
    
    return nodes
```

```python
def nodes_dictionary(df):    
    nodes_dict = {}
    for i in range(len(df)):
        u = df.iloc[i, 0]
        v = df.iloc[i, 1]

        if u not in nodes_dict:
            nodes_dict[u] = [v]
        else:
            nodes_dict[v] = [u]
            
    return nodes_dict
```

```python
def getMatrixSize(df):
    size = 0
    for i in range(len(df)):
        u = df.iloc[i, 0]
        v = df.iloc[i, 1]
        
        if size < u:
            size = u
        if size < v:
            size = v
    
    return size
```

```python
nodes_dict = nodes_dictionary(df)
nodes = getNodes(df)
size = getMatrixSize(df)
```
